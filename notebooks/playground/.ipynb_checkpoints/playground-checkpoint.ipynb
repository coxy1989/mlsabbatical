```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pandas - Basics
# http://pandas.pydata.org/pandas-docs/stable/basics.html#basics

dat = np.random.randn(10,4)
index = pd.date_range('1/1/2018', periods=10)
cols = ['a', 'b', 'c', 'd']
df = pd.DataFrame(dat, index=index, columns=cols)

df
df.head()
df.tail()
df.describe()
```

```python
# Pandas - Series
# http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dsintro

# series from scalar
s = pd.Series(5.0, index=['a','b','c'])

# series with custom labels
s = pd.Series([1,2,3], index=['a','b','c'])

# series from dict
s = pd.Series({'a' : 1, 'b' : 2})
s = pd.Series({'a' : 1, 'b' : 2, 'c' : 3}, index= ['a', 'b', 'f'])

# slicing
s[0 : 2]

# indexing
s['a'] = 10
s.get('j')

# vectorized operations
s1 = pd.Series(np.random.rand(5))
s2 = pd.Series(np.random.rand(5))
s1 + s2
s1 @ s2
np.exp(s1)


'''
A key difference between Series and ndarray is that operations between Series automatically
align the data based on label. Thus, you can write computations without giving consideration
to whether the Series involved have the same labels.
'''

# Pandas..

s1 = pd.Series({'a': 1, 'b' : 2, 'c' :3})
s2 = pd.Series({'b' : 20, 'c' :30})
s1 + s2

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([1, 2])

s1 + s2

# Numpy..

s1 = np.array([1,2,3])
s2 = np.array([1,2])

# error..
#s1 + s2

```

```python
# Pandas - DataFrame
# http://pandas.pydata.org/pandas-docs/stable/dsintro.html#dsintro

d = {'one': pd.Series([1,2,3], index=['a','b','c']),
     'two': pd.Series([4,5,6,7], index=['a','b','c', 'd'])}

pd.DataFrame(d)
pd.DataFrame(d,index=['b','c'])
pd.DataFrame(d,index=['b','c'],columns=['one'])

pd.DataFrame(d).index
pd.DataFrame(d).columns

```

```python

tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(x="time", y="firing_rate", col="align",
            hue="choice", size="coherence", style="choice",
            facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots);
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", col="region",
            hue="event", style="event",
            kind="line", data=fmri);
```

```python
sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",
           data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="swarm", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="bar", data=tips);
```

```python
iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="petal_length", data=iris);
```
