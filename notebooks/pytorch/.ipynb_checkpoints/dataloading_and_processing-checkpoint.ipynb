```python
def plot_face(img, landmarks):
    plt.imshow(img)
    #lm = row[1:].values.reshape(-1, 2)
    plt.scatter(landmarks[:,0], landmarks[:,1], marker='.')
```

```python
class FaceLandmarksDataset(torch.utils.data.Dataset):
    
    def __init__(self, csv_path, data_dir, transform=None):
        self.landmarks_df = pd.read_csv(csv_path)
        self.data_dir = data_dir
        self.transform = transform
        
    def __len__(self):
        return len(self.landmarks_df)
    
    def __getitem__(self, idx):
        row = self.landmarks_df.iloc[idx]
        image_name = row['image_name']
        image = skimage.io.imread(f'{self.data_dir}{image_name}')
        landmarks = row[1:].values.reshape(-1, 2)
        sample = {'image':image, 'landmarks' : landmarks.astype(float)}
        if self.transform:
            sample = self.transform(sample)
        return sample
```

```python
def to_tensor(sample):
    image, landmarks = sample['image'], sample['landmarks']
    # numpy image: H x W x C
    # torch image: C * H * W
    image = image.transpose((2, 0, 1))
    return {'image' : torch.from_numpy(image),
            'landmarks' : torch.from_numpy(landmarks)}
    
def random_crop(out_size, sample):
    image, landmarks = sample['image'], sample['landmarks']
    # assume numpy image: H x W x C
    h, w = image.shape[:2]
    top = np.random.randint(0, h - out_size)
    left = np.random.randint(0, w - out_size)
    landmarks = landmarks - [left, top]
    
    image = image[top : top + out_size,
                  left : left + out_size]
    
    return {'image' : image,
           'landmarks' : landmarks}
    
def rescale(out_size, sample):
    #assume numpy image: H x W x C
    image, landmarks = sample['image'], sample['landmarks']
    h, w = image.shape[:2]
    if h > w:
        new_h, new_w = out_size * h / w, out_size
    else:
        new_h, new_w = out_size, out_size * w / h
    image = skimage.transform.resize(image, (int(new_h), int(new_w)), mode='constant')
    landmarks = landmarks * [new_w / w, new_h / h]
    return {'image' : image,
            'landmarks' : landmarks}
    
tfm = torchvision.transforms.Compose([ft.partial(rescale, 256),
                                      ft.partial(random_crop, 224),
                                      to_tensor])
```

```python
csv_path = './data/faces/face_landmarks.csv'
data_path = './data/faces/'
fl_dataset = FaceLandmarksDataset(csv_path, data_path) #transform=tfm)
sample = fl_dataset[3]
cropped = random_crop(224, sample)
scaled = rescale(256, sample)
composed = random_crop(224, scaled)

plot_face(cropped['image'], cropped['landmarks'])
plt.show()
plot_face(scaled['image'], scaled['landmarks'])
plt.show()
plot_face(composed['image'], composed['landmarks'])
```
