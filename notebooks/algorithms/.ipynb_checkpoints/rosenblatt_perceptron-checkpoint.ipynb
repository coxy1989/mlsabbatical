```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(10,10)})
```

[Perceptrons](https://en.wikipedia.org/wiki/Perceptron) were cooked up by by Frank Rosenblatt  in 1957.

## Model

We will try and find a seperating hyperplane by minimizing the distance of misclassified points to the decision boundry.

Define a hyperplane:

$\beta_0 + \beta^T x = 0$

Our goal is to minimize:

$D(\beta, \beta_0) = - \sum_{i \in \mathcal{M}} y_i (x_i^T \beta + \beta_0)$

*Where* 
- $y \in \{ -1, 1\}$
- $\mathcal{M}$ indexes the set of misclassified points

Differentiating with respect to $\beta$ and $\beta_0$ yields:

$\frac{\partial D(\beta, \beta_0)}{\partial \beta} = - \sum\limits_{i \in \mathcal{M}} y_i x_i$

$\frac{\partial D(\beta, \beta_0)}{\partial \beta_0} = - \sum\limits_{i \in \mathcal{M}} y_i$

With these partials in hand, we can use a first order optimization algorithm to solve numerically for $\beta$ and $\beta_0$

### Generate some data

```python
# two liearly seperatble classes
X = np.random.uniform(0, 1, (200, 2))
y = (X[:,0] > X[:,1]).reshape(-1, 1)
X[y.T[0]] = X[y.T[0]] + [0.1, -0.1]

#plot
scatter_df = pd.concat([pd.DataFrame(X), pd.Series(y.T[0])], axis=1)
scatter_df.columns = ['x1', 'x2', 'class']
sns.scatterplot(x='x1', y='x2', hue='class', data=scatter_df)
```

### Gradient Descent

```python
def batch_gd(n_iter, initial_beta):
    i = 0
    beta = initial_beta
    intercept = np.repeat(1, X.shape[0]).reshape(-1,1)
    d_matrix = np.hstack([intercept, X])
    lr = 0.001
    while i < n_iter:
        preds = d_matrix @ beta
        misclassified = (preds > 0) != y
        #  binary y values to {-1, +1}
        y_weights = ((y * 2) - 1)
        #  correctly classified xs to zero
        xs = misclassified * d_matrix
        # negated sum of product of ys and xs for misclassified training examples
        partial_beta = - (y_weights * xs).mean(axis=0).reshape(-1,1)
        beta = beta - (lr * partial_beta)
        i += 1
    return beta
```

### Decision Boundry

```python
def plot_classifier(beta):
    ax = sns.scatterplot(x='x1', y='x2', hue='class', data=scatter_df)
    xmin, xmax = ax.get_xlim()
    ymin, ymax = ax.get_ylim()
    xx1 = np.linspace(xmin, xmax, 500)
    xx2 = np.linspace(ymin, ymax, 500)
    xx1, xx2 = np.meshgrid(xx1, xx2)
    intercept = np.repeat(1, xx1.ravel().shape[0])
    X_grid_d_matrix = np.vstack([intercept ,xx1.ravel(), xx2.ravel()]).T
    zz = (X_grid_d_matrix @ beta > 0).reshape(xx1.shape)
    plt.contourf(xx1, xx2, zz, cmap='coolwarm', alpha=0.25)
```

When the data is seperable, there are infinitely many solutions, which one is found depends on the starting value we choose for our optimization:

```python
n_iter = (10 ** 4)
initial_beta = np.repeat(0, 3).reshape(-1, 1)
beta = batch_gd(n_iter, initial_beta),
plot_classifier(beta)
```

```python
initial_beta = np.repeat(-0.1, 3).reshape(-1, 1)
beta = batch_gd(n_iter, initial_beta),
plot_classifier(beta)
```

```python
initial_beta = np.repeat(0.1, 3).reshape(-1, 1)
beta = batch_gd(n_iter, initial_beta),
plot_classifier(beta)
```
