```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(10,10)})
```

### Algorithm

```python
def stratified_random_start(n, k):
    quot = n // k
    mod = n % k
    quot_seq = [np.arange(0, k) for _ in range(0, quot)]
    mod_seq = (np.arange(0, mod))
    ret = np.concatenate([*quot_seq, mod_seq], axis=0)
    np.random.shuffle(ret)
    return ret
    
def distance(X, centroid):
    return np.sqrt(((centroid - X) ** 2).sum(axis=1))   

def k_means(X, k):
    n = X.shape[0]
    assigns = stratified_random_start(n, k)
    centroids = [np.repeat(0, 2).reshape(1, -1) for _ in range(0, k)]
    l = 0
    while l < 10:
        next_centroids = []
        distances = []
        for i, c in enumerate(centroids):
            pts = X[assigns == i]
            nc = pts.mean(axis=0) if pts.shape[0] > 0 else c
            next_centroids.append(nc)
            dist = distance(X, nc)
            distances.append(dist)
            dist_table = np.vstack(distances).T
        assigns = pd.DataFrame(dist_table).idxmin(axis=1).values
        centroids = next_centroids
        l += 1
    return assigns, centroids
```

```python
n = 1000
k = 4
class_1 = np.random.normal([3, 3], 1, (n//k, 2))
class_2 = np.random.normal([3, -3], 1,(n//k, 2))
class_3 = np.random.normal([-3,-3], 1, (n//k, 2))
class_4 = np.random.normal([-3, 3], 1, (n//k, 2))

df = pd.concat([pd.DataFrame(class_1),
                pd.DataFrame(class_2),
                pd.DataFrame(class_3),
                pd.DataFrame(class_4)], axis=0, ignore_index=True)
df.columns = ['x1', 'x2']

assigns, centroids = k_means(df, k)
scatter_df = pd.concat([df, pd.Series(assigns).rename('class')], axis=1)
ax = sns.scatterplot(x='x1', y='x2', hue='class', data=scatter_df,palette=sns.color_palette(n_colors=len(np.unique(assigns))))
for c in centroids:
    _ = ax.text(c.x1, c.x2, 'X', color='black')
```

```python
n = 10000
k = 20
dfs = [pd.DataFrame(np.random.normal([np.random.uniform(0, 15), np.random.uniform(0, 15)], 1, (n//k, 2))) for i in range(0, k)]

df = pd.concat(dfs, axis=0, ignore_index=True)
df.columns = ['x1', 'x2']

assigns, centroids = k_means(df, k)
scatter_df = pd.concat([df, pd.Series(assigns).rename('class')], axis=1)
ax = sns.scatterplot(x='x1', y='x2', hue='class', data=scatter_df,palette=sns.color_palette(n_colors=len(np.unique(assigns))))
for c in centroids:
    _ = ax.text(c.x1, c.x2, 'X', color='black')
```
